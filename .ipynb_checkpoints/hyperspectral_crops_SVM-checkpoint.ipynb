```python
import numpy as np 
import pandas as pd

import os

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(font_scale=1.5)


import time


from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# load data
f = "GHISACONUS_2008_001_speclib.csv" #data file
df = pd.read_csv(f)
```

```python
df.head()
```

```python
df["Crop"].unique()
```

```python
df["Stage"].unique()
```

```python
df["Class"] = df["Crop"] + "_" + df["Stage"]
```

```python
df["Class"].unique()
```

```python
df["Class"].nunique()
```

```python
# wavelength columns
wl_list = []

for c in df.columns:
    if c[0]=='X': wl_list.append(c)
```

```python
X = df[wl_list]
```

```python
X.head()
```

```python
X1 = StandardScaler().fit_transform(X)
```

```python
Y = df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X1,Y,test_size=0.20, random_state = 0)
```

```python
model = SVC( C = 1e4,  gamma = 1e-4, tol =1e-6 )
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, pred)
accuracy
```

```python
df_result = pd.DataFrame()
df_result["Ground Truth"] = y_test
df_result["Prediction"] = pred
df_result["Boolean"] = df_result["Ground Truth"] == df_result["Prediction"]
df_result.sort_values(by = "Ground Truth", inplace=True)

df_result[df_result["Boolean"] == False].drop(columns=["Boolean"]).sort_values(by = "Ground Truth").to_excel("misidentification.xlsx")
```

```python
cf_matrix = confusion_matrix(df_result["Ground Truth"],df_result["Prediction"])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

axis_labels = df_result["Ground Truth"].unique()

s = sns.heatmap(cf_matrix, xticklabels= axis_labels, yticklabels= axis_labels, annot = True,  annot_kws={'size':10},cmap='Blues')
s.tick_params(labelsize = 10)
show_annot_array = cf_matrix > 0

for text, show_annot in zip(ax.texts, show_annot_array.ravel()):
    text.set_visible(show_annot)

plt.xlabel("Predicted",fontsize = 18)
plt.ylabel("Actual",fontsize = 18)


plt.savefig("confusion_matrix.png", dpi = 300, bbox_inches='tight')
```

```python
cf_matrix_n = cf_matrix.astype('float')/cf_matrix.sum(axis = 1)[:,np.newaxis]


fig, ax = plt.subplots(1, 1, figsize=(10, 8))

axis_labels = df_result["Ground Truth"].unique()

s = sns.heatmap(cf_matrix_n, xticklabels= axis_labels, yticklabels= axis_labels, annot = True,  annot_kws={'size':6},cmap='Blues',fmt = '.2f' )
s.tick_params(labelsize = 10)
show_annot_array = cf_matrix > 0

for text, show_annot in zip(ax.texts, show_annot_array.ravel()):
    text.set_visible(show_annot)

plt.xlabel("Predicted",fontsize = 18)
plt.ylabel("Actual",fontsize = 18)


plt.savefig("confusion_matrix_n.png", dpi = 300, bbox_inches='tight')
```

```python
from sklearn.metrics import precision_recall_fscore_support as score
precision, recall, fscore, support = score(df_result["Ground Truth"],df_result["Prediction"])

df = pd.DataFrame()
df["Gemstone"] = df_result["Ground Truth"].unique()
df["Precision"] = precision
df["Recall"] = recall
df["F-score"] = fscore
df["Support"] = support
df.to_excel("score.xlsx")
df
```
